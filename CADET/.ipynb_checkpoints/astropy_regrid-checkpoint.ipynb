```python
def regrid(fname, scale):
    file = fits.open(fname)
    wcs0 = WCS(file[0].header)
    data0 = file[0].data
    shape = data0.shape[0]

    # CROP
    x0 = shape / 2
    size = 128 * scale
    cutout = Cutout2D(data0, (x0, x0), (size, size), wcs=wcs0)
    data, wcs = cutout.data, cutout.wcs

    ra, dec = wcs.wcs_pix2world(np.array([[63, 63]]),0)[0]
    
    # REGRID
    factor = size // 128
    data_new = data.reshape(128, factor, 128, factor).mean(-1).mean(1)
    
    wcs.wcs.cdelt[0] = wcs.wcs.cdelt[0] * factor
    wcs.wcs.cdelt[1] = wcs.wcs.cdelt[1] * factor
    
    wcs.wcs.crval[0] = ra
    wcs.wcs.crval[1] = dec
    
    wcs.wcs.crpix[0] = 64 / factor
    wcs.wcs.crpix[1] = 64 / factor

    return data_new, wcs
```

```python
data, wcs = regrid("NGC4649.fits", 1)

ccd = CCDData(data, unit="adu", wcs=wcs)
ccd.write("NGC4649_regrid.fits", overwrite=True)
```

```python
data, wcs = regrid("NGC5813.fits", 1)

ccd = CCDData(data, unit="adu", wcs=wcs)
ccd.write("NGC5813_regrid.fits", overwrite=True)
```

```python
def rebin(arr, new_shape):
    shape = (new_shape[0], arr.shape[0] // new_shape[0],
             new_shape[1], arr.shape[1] // new_shape[1])
    return arr.reshape(shape).mean(-1).mean(1)
```

```python
data_new = rebin(data, (128,128))
```
